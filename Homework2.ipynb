Bài 8: Giải hệ phương trình tuyến tính:
ax + by = m
cx + dy = n

```python
print("He phuong trinh tuyen tinh: ax+by=m va cx+dy=n")
a = int(input("He so a:"))
b = int(input("He so b:"))
m = int(input("So m:"))
c = int(input("He so c:"))
d = int(input("He so d:"))
n = int(input("So n:"))
b1 = b*c
d1 = d*a
m1 = m*c
n1 = n*a
y = (m1-n1)/(b1-d1)
x1 = (m-(b*y))/a
x2 = (n-(d*y))/c
if x1==x2:
    print("Nghiem cua he phuong trinh lan luot la:x y =",x1,y)
else: 
    print("He vo nghiem")
```
